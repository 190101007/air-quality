```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, silhouette_score, accuracy_score
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans, HDBSCAN
from sklearn.cluster import DBSCAN
from sklearn.model_selection import cross_val_score
from sklearn.multiclass import OneVsRestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import reciprocal, randint
from sklearn.neighbors import NearestNeighbors
import plotly.express as px

```

```python
df = pd.read_csv("updated_pollution_dataset.csv")
```

```python
df.info()
```

```python
df.describe().T
```

```python
X = df.drop(["Air Quality"], axis=1)
y = df["Air Quality"]
```

```python
correlation_matrix = X.corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, linewidth=.7)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(data=df[["Temperature", "Humidity", "PM2.5", "PM10" , "NO2", "SO2", "CO", "Proximity_to_Industrial_Areas"]])
plt.title("tm kutu grafikleri")
plt.xlabel("deerler")
plt.show()
```

```python
target_counts = y.value_counts()
print(target_counts)
print(target_counts / len(df) * 100)  # Yzdelik dalm
```

```python
plt.figure(figsize=(7,4))
sns.countplot(data=df, x='Air Quality')
plt.title('Hedef Deiken Dalm')
plt.show()
```

```python
quality = [['Hazardous', 'Poor', 'Moderate', 'Good']]
```

```python
y_ordinal = OrdinalEncoder(categories=quality)
y = y_ordinal.fit_transform(df[["Air Quality"]])
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns, index=X.index)
X_scaled_df.head()
```

```python
outlier_mask = (np.abs(X_scaled) > 6).any(axis=1)
combined_mask = ((np.abs(X_scaled) > 3).any(axis=1)) # | mask_2 | mask_3
X_scaled = X_scaled[~combined_mask]
y = y[~combined_mask]
```

```python
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
smote = SMOTE(random_state=42)
x_balanced, y_balanced = smote.fit_resample(X_scaled_df, y)
```

```python
print(x_balanced)
print("lk 20 eleman:", y_balanced[:20])
```

```python
pca = PCA(n_components=4)
X_reduced = pca.fit_transform(x_balanced)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_reduced, y_balanced, test_size=0.2, random_state=42)
```

```python
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y_balanced, cmap='viridis')  # y_balanced varsa etiket olarak
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('PCA ile Grselletirme')
plt.colorbar()
plt.show()
```

```python
# DataFrame
df_tsne = pd.DataFrame({
    'tsne1': X_reduced[:, 0],
    'tsne2': X_reduced[:, 1],
    'tsne3': X_reduced[:, 2],
    'label': y_balanced
})

# Etkileimli 3B scatter plot
fig = px.scatter_3d(df_tsne, x='tsne1', y='tsne2', z='tsne3',
                    color=y_balanced,
                    title='3 Boyutlu t-SNE Grselletirmesi',
                    labels={'tsne1': 't-SNE Boyut 1',
                            'tsne2': 't-SNE Boyut 2',
                            'tsne3': 't-SNE Boyut 3'},
                    color_continuous_scale='viridis')

fig.update_layout(
    width=500,  # piksel cinsinden genilik
    height=500, # piksel cinsinden ykseklik
    autosize=False  # otomatik boyutlandrmay kapat
)


fig.show()
```

LOGISTIC REGRESSION

```python
reg = OneVsRestClassifier(LogisticRegression(max_iter=5000, penalty='elasticnet', solver='saga', l1_ratio=0.5))
model = reg.fit(X_train, y_train)

#PREDICT
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

#ROC AUC iin olaslk predict
y_pred_proba_train = model.predict_proba(X_train)
y_pred_proba_test = model.predict_proba(X_test)
```

```python
train_accuracy = accuracy_score(y_train, y_pred_train)
test_accuracy = accuracy_score(y_test, y_pred_test)
print(f"Eitim Verisi Doruluu (Training Accuracy): {train_accuracy:.4f}")
print(f"Test Verisi Doruluu (Testing Accuracy): {test_accuracy:.4f}")
```

```python
train_roc_auc = roc_auc_score(y_train, y_pred_proba_train, multi_class='ovr')
test_roc_auc = roc_auc_score(y_test, y_pred_proba_test, multi_class='ovr')
print(f"Eitim Verisi ROC AUC Skoru: {train_roc_auc:.4f}")
print(f"Test Verisi ROC AUC Skoru: {test_roc_auc:.4f}")
```

```python
k_accuracy_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')
k_roc_auc_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='roc_auc_ovr')
print(f"5-Fold Cross-Validation ROC AUC (OvR) Skorlar: {np.round(k_roc_auc_scores, 4)}")
print(f"Ortalama ROC AUC Skoru: {k_roc_auc_scores.mean():.4f}")
print(f"Accuracy Cross-Validation: {np.round(k_accuracy_scores, 4)}")
```

```python
print("\n--- Snflandrma Raporu ---")
print(classification_report(y_test, y_pred_test))
```

```python
print("--- Karklk Matrisi (Confusion Matrix) ---")
print(confusion_matrix(y_test, y_pred_test))
```

KMEANS

```python
#elbow metodu
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, random_state=42)
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1, 11), wcss, marker="o")
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method for KMeans")
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
clusters_kmeans = kmeans.fit_predict(X_train)

# PCA ile grselletirme
pca_vis = PCA(n_components=3)
X_vis = pca_vis.fit_transform(X_train)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_vis[:, 0], X_vis[:, 1], c=clusters_kmeans, cmap="viridis", s=50)
plt.title("KMeans Clusters (PCA 2D)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
X_tsne = TSNE(n_components=3, random_state=42).fit_transform(X_train)
```

```python
plt.scatter(X_tsne[:,0], X_tsne[:,1], c=clusters_kmeans, cmap="viridis", s=50)
plt.title("KMeans Clusters (t-SNE)")
plt.show()
```

```python
# DataFrame
df_tsne = pd.DataFrame({
    'tsne1': X_tsne[:, 0],
    'tsne2': X_tsne[:, 1],
    'tsne3': X_tsne[:, 2],
    'label': y_train
})

# Etkileimli 3B scatter plot
fig = px.scatter_3d(df_tsne, x='tsne1', y='tsne2', z='tsne3',
                    color=clusters_kmeans,
                    title='3 Boyutlu t-SNE Grselletirmesi',
                    labels={'tsne1': 't-SNE Boyut 1',
                            'tsne2': 't-SNE Boyut 2',
                            'tsne3': 't-SNE Boyut 3'},
                    color_continuous_scale='viridis')

fig.update_layout(
    width=500,  # piksel cinsinden genilik
    height=500, # piksel cinsinden ykseklik
    autosize=False  # otomatik boyutlandrmay kapat
)


fig.show()
```

DBSCAN

```python
min_samples=4
neigh = NearestNeighbors(n_neighbors=min_samples)
distances, indices = neigh.fit(X_train).kneighbors(X_train)

# En yakn min_samples'inci komunun mesafelerini sralama
distances = np.sort(distances[:, min_samples - 1], axis=0)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(distances)
plt.xlabel("Veri Noktalar (Sralanm)")
plt.ylabel(f"{min_samples}. En Yakn Komuya Olan Mesafe")
plt.title("DBSCAN in Dirsek Yntemi (Epsilon Belirleme)")
plt.grid(True)
plt.show()
```

```python
dbscan = DBSCAN(eps=0.6, min_samples=6)
clusters = dbscan.fit_predict(X_train)
```

```python
#PCA GRSELLETRME
plt.figure(figsize=(8,6))
plt.scatter(X_vis[:, 0], X_vis[:, 1], c=clusters, cmap="viridis", s=50)
plt.title("DBSCAN Clusters (PCA 2D)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
