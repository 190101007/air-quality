```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, silhouette_score, accuracy_score, mean_absolute_error, f1_score, precision_score, recall_score
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.multiclass import OneVsRestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import reciprocal, randint
from sklearn.neighbors import NearestNeighbors
import plotly.express as px
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv("updated_pollution_dataset.csv")
```

```python
X = df.drop(["Air Quality"], axis=1)
y = df["Air Quality"]
```

```python
encoder = OneHotEncoder(drop=None)  # drop='first' is optional
y_encoded = encoder.fit_transform(df[["Air Quality"]])

# Etiketlere geri dnmek istiyorsan:
y_labels = np.argmax(y_encoded, axis=1)
y = y_labels
y = np.array(y).ravel()

```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns, index=X.index)
X_scaled_df.head()
```

```python
#outlier_mask = (np.abs(X_scaled) > 6).any(axis=1)
combined_mask = ((np.abs(X_scaled) > 3).any(axis=1)) # | mask_2 | mask_3
X_scaled = X_scaled[~combined_mask]
y = y[~combined_mask]
```

```python
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
smote = SMOTE(random_state=42)
x_balanced, y_balanced = smote.fit_resample(X_scaled_df, y)
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x=y_balanced)
plt.title('Hedef Deiken Dalm')
plt.show()
```

```python
x_balanced.head(5)
```

```python
print("lk 20 eleman y:", y_balanced[:20])
```

```python
pca = PCA(n_components=0.95)
X_reduced = pca.fit_transform(x_balanced)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_reduced, y_balanced, test_size=0.2, random_state=42)
```

```python
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y_balanced, cmap='viridis')  # y_balanced etiket olarak
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('PCA ile Grselletirme')
plt.colorbar()
plt.show()
```

```python
# DataFrame
df_tsne = pd.DataFrame({
    'tsne1': X_reduced[:, 0],
    'tsne2': X_reduced[:, 1],
    'tsne3': X_reduced[:, 2],
    'label': y_balanced
})

# Etkileimli 3B scatter plot
fig = px.scatter_3d(df_tsne, x='tsne1', y='tsne2', z='tsne3',
                    color=y_balanced,
                    title='3 Boyutlu t-SNE Grselletirmesi',
                    labels={'tsne1': 't-SNE Boyut 1',
                            'tsne2': 't-SNE Boyut 2',
                            'tsne3': 't-SNE Boyut 3'},
                    color_continuous_scale='viridis')

fig.update_layout(
    width=500,  # piksel cinsinden genilik
    height=500, # piksel cinsinden ykseklik
    autosize=False  # otomatik boyutlandrmay kapat
)

fig.write_image("3d_tsne_gorseli.png")


fig.show()
```

LOGISTIC REGRESSION

```python
param_grid =[ {
    'estimator__C': [0.01, 0.1, 1, 10, 100],
    'estimator__penalty': ['l1', 'l2'],
    'estimator__solver': ['saga']
},
{'estimator__C': [0.01, 0.1, 1, 10, 100],
     'estimator__penalty': ['elasticnet'],
     'estimator__solver': ['saga'],
     'estimator__l1_ratio': [0.2, 0.5, 0.8]} ]

reg = OneVsRestClassifier(LogisticRegression(max_iter=5000))
grid_search = GridSearchCV(reg, param_grid, cv=5, scoring='roc_auc_ovr', n_jobs=-1)
grid_search.fit(X_train, y_train)
```

```python
print("En iyi hiperparametreler:", grid_search.best_params_)
print("En iyi ROC AUC skoru:", grid_search.best_score_)

# En iyi modeli kullann
model = grid_search.best_estimator_
```

```python
#PREDICT
y_pred_train_lr = model.predict(X_train)
y_pred_test_lr = model.predict(X_test)

#ROC AUC iin olaslk predict
y_pred_proba_train_lr = model.predict_proba(X_train)
y_pred_proba_test_lr = model.predict_proba(X_test)
```

```python
train_accuracy_lr = accuracy_score(y_train, y_pred_train_lr)
test_accuracy_lr = accuracy_score(y_test, y_pred_test_lr)
print(f"Eitim Verisi Doruluu (Training Accuracy): {train_accuracy_lr:.4f}")
print(f"Test Verisi Doruluu (Testing Accuracy): {test_accuracy_lr:.4f}")
```

```python
train_roc_auc = roc_auc_score(y_train, y_pred_proba_train_lr, multi_class='ovr')
test_roc_auc_lr = roc_auc_score(y_test, y_pred_proba_test_lr, multi_class='ovr')
print(f"Eitim Verisi ROC AUC Skoru: {train_roc_auc:.4f}")
print(f"Test Verisi ROC AUC Skoru: {test_roc_auc_lr:.4f}")
```

```python
print("\n--- Snflandrma Raporu ---")
cl_lr = classification_report(y_test, y_pred_test_lr)
print(cl_lr)
```

```python
cm_lr = confusion_matrix(y_test, y_pred_test_lr)
plt.figure(figsize=(4,3))
sns.heatmap(cm_lr, annot=True, fmt='d', xticklabels=np.unique(y), yticklabels=np.unique(y))
plt.title('Karklk Matrisi (Confusion Matrix)')
plt.xlabel('Tahmin Edilen Snf')
plt.ylabel('Gerek Snf')
plt.show()
```

```python
mae_lr = mean_absolute_error(y_test, y_pred_test_lr)

print(f"MAE: {mae_lr:.4f}")
```

KMEANS

```python
#elbow metodu
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, random_state=42)
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1, 11), wcss, marker="o")
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method for KMeans")
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
clusters_kmeans = kmeans.fit_predict(X_train)

# PCA ile grselletirme
pca_vis = PCA(n_components=3)
X_vis = pca_vis.fit_transform(X_train)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_vis[:, 0], X_vis[:, 1], c=clusters_kmeans, cmap="viridis", s=50)
plt.title("KMeans Clusters (PCA 2D)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
skor_kmeans = silhouette_score(X_train, clusters_kmeans)
print(skor_kmeans)
```

```python
X_tsne = TSNE(n_components=3, random_state=42).fit_transform(X_train)
```

```python
plt.scatter(X_tsne[:,0], X_tsne[:,1], c=clusters_kmeans, cmap="viridis", s=50)
plt.title("KMeans Clusters (t-SNE)")
plt.show()
```

```python
# DataFrame
df_tsne = pd.DataFrame({
    'tsne1': X_tsne[:, 0],
    'tsne2': X_tsne[:, 1],
    'tsne3': X_tsne[:, 2],
    'label': y_train
})

# Etkileimli 3B scatter plot
fig = px.scatter_3d(df_tsne, x='tsne1', y='tsne2', z='tsne3',
                    color=clusters_kmeans,
                    title='3 Boyutlu t-SNE Grselletirmesi',
                    labels={'tsne1': 't-SNE Boyut 1',
                            'tsne2': 't-SNE Boyut 2',
                            'tsne3': 't-SNE Boyut 3'},
                    color_continuous_scale='viridis')

fig.update_layout(
    width=500,  # piksel cinsinden genilik
    height=500, # piksel cinsinden ykseklik
    autosize=False  # otomatik boyutlandrmay kapat
)


fig.show()
```

DBSCAN

```python
min_samples=4
neigh = NearestNeighbors(n_neighbors=min_samples)
distances, indices = neigh.fit(X_train).kneighbors(X_train)

# En yakn min_samples'inci komunun mesafelerini sralama
distances = np.sort(distances[:, min_samples - 1], axis=0)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(distances)
plt.xlabel("Veri Noktalar (Sralanm)")
plt.ylabel(f"{min_samples}. En Yakn Komuya Olan Mesafe")
plt.title("DBSCAN in Dirsek Yntemi (Epsilon Belirleme)")
plt.grid(True)
plt.show()
```

```python
dbscan = DBSCAN(eps=1.2, min_samples=6)
clusters = dbscan.fit_predict(X_train)
```
